# Feature selection using linear regression



### Feature selection on the data from keybert

```python
from sklearn.datasets import make_regression
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
import operator
 
# feature selection
def select_features(X_train, y_train, X_test):
	# configure to select all features
	fs = SelectKBest(score_func=f_regression, k='all')
	# learn relationship from training data
	fs.fit(X_train, y_train)
	# transform train input data
	X_train_fs = fs.transform(X_train)
	# transform test input data
	X_test_fs = fs.transform(X_test)
	return X_train_fs, X_test_fs, fs
 
# feature selection
X_train_fs, X_test_fs, fs = select_features(x_train_keybert, y_train, x_test_keybert)

all_features = {}
for i in range(len(fs.scores_)):
  all_features[list(x_train_keybert.columns)[i]] = fs.scores_[i]

sorted_d = dict(sorted(all_features.items(), key=operator.itemgetter(1),reverse=True))
# print(sorted_d)
# write to csv
feature_scores_keybert = pd.DataFrame([sorted_d]) 
feature_scores_keybert.to_csv(r'/content/drive/My Drive/BT4222 group project/final_data/feature_scores_keybert.csv', index = False)
```

### Feature selection on the data from topic modelling

```python
# feature selection
X_train_fs, X_test_fs, fs = select_features(x_train_topics, y_train, x_test_topics)

all_features = {}
for i in range(len(fs.scores_)):
  all_features[list(x_train_topics.columns)[i]] = fs.scores_[i]

sorted_d = dict(sorted(all_features.items(), key=operator.itemgetter(1),reverse=True))
sorted_d
# write to csv
feature_scores_topics = pd.DataFrame([sorted_d]) 
feature_scores_topics.to_csv(r'/content/drive/My Drive/BT4222 group project/final_data/feature_scores_topics.csv', index = False)
```
